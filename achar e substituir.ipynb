```python
'''
converte matriz de espaços em .txt para .npy
'''
import numpy as np

# arquivo de entrada e de saída
input_path = "matrix.txt"
output_path = "matrix.npy"

# escolher o separador
separador = " "
#separador = ", "

# lê os dados dos .txt na memória
with open(input_path) as f:
    matrix = [[float(y) for y in x.strip().split()] for x in f.readlines()]

# salva em npy
np.save(output_path, matrix)


'''
converte matrix.npy para matrix.txt
'''
import numpy as np

# arquivo de entrada e de saída
input_path = "matrix.npy"
output_path = "matrix.txt"

# escolher o separador
separador = " "
#separador = ", "

# abre o npy
matrix = np.load(input_path)

# pega o tamanho da matrix (que é o tamanho do dataset)
n = len(matrix)

# escreve em .txt
f = open(output_path, "w+")
for i in range(n):
    for j in range(n):
        # escreve o elemento com o separador
        print(matrix[i][j], end=separador, file=f)
    # pula linha
    print(end="\n", file=f)
# fecha arquivo
f.close()
```
